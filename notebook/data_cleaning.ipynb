```python

# =====================================================
# HOTEL BOOKING DATA CLEANING PROJECT
# =====================================================

# =====================================================
# 1. IMPORT LIBRARIES
# Purpose: Import required libraries
# =====================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder


# =====================================================
# 2. LOAD DATASET
# Purpose: Read hotel_booking.csv file
# =====================================================

df = pd.read_csv("../datasets/hotel_booking.csv")
print("Dataset Loaded Successfully")

print("Rows and Columns :", df.shape)


# =====================================================
# 3. BASIC DATASET INFORMATION
# Purpose: Understand dataset structure
# =====================================================

print("\n========== DATASET INFORMATION ==========\n")

df.info()

print("\n========== STATISTICAL SUMMARY ==========\n")

print(df.describe())

print("\n========== MISSING VALUES ==========\n")

print(df.isnull().sum())


# =====================================================
# 4. HANDLE MISSING VALUES
# Purpose: Fill null values
# =====================================================

df["country"] = df["country"].fillna(
    df["country"].mode()[0]
)

df["children"] = df["children"].fillna(0)

df["agent"] = df["agent"].fillna(0)

df["company"] = df["company"].fillna(0)

print("\nMissing Values Removed")


# =====================================================
# 5. REMOVE DUPLICATES
# Purpose: Delete duplicate rows
# =====================================================

print("\nDuplicates Before :",
      df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicates After :",
      df.duplicated().sum())


# =====================================================
# 6. FIX DATA TYPES
# Purpose: Convert incorrect datatypes
# =====================================================

df["children"] = df["children"].astype(int)

df["agent"] = df["agent"].astype(int)

df["company"] = df["company"].astype(int)

df["reservation_status_date"] = pd.to_datetime(
    df["reservation_status_date"]
)

print("\nData Types Corrected")


# =====================================================
# 7. STANDARDIZE COLUMN NAMES
# Purpose: Make column names uniform
# =====================================================

df.columns = (

    df.columns

    .str.lower()

    .str.strip()

    .str.replace(" ", "_")

    .str.replace("-", "_")

)

print("\nColumn Names Standardized")


# =====================================================
# 8. FORMAT CATEGORICAL VALUES
# Purpose: Standardize text values
# =====================================================

categorical_columns = [

    "hotel",

    "meal",

    "market_segment",

    "distribution_channel",

    "customer_type"

]

for col in categorical_columns:

    df[col] = (

        df[col]

        .astype(str)

        .str.strip()

        .str.title()

    )

print("\nCategorical Values Standardized")


# =====================================================
# 9. CHECK INCONSISTENT VALUES
# Purpose: Remove negative values
# =====================================================

cols = [

    "adr",

    "adults",

    "children",

    "babies"

]

for col in cols:

    df = df[df[col] >= 0]

print("\nInconsistent Values Removed")


# =====================================================
# 10. VALIDATE RELATIONSHIPS
# Purpose: At least one guest should exist
# =====================================================

df = df[

    (

        df["adults"]

        + df["children"]

        + df["babies"]

    ) > 0

]

print("\nRelationship Validation Completed")


# =====================================================
# 11. DETECT AND REMOVE OUTLIERS
# Purpose: Remove extreme values
# =====================================================

outlier_columns = [

    "lead_time",

    "adr",

    "stays_in_week_nights",

    "stays_in_weekend_nights"

]

for col in outlier_columns:

    Q1 = df[col].quantile(0.25)

    Q3 = df[col].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)

    upper = Q3 + (1.5 * IQR)

    df = df[

        (df[col] >= lower)

        &

        (df[col] <= upper)

    ]

print("\nOutliers Removed")


# =====================================================
# 12. ENCODE CATEGORICAL DATA
# Purpose: Convert text into numbers
# =====================================================

le = LabelEncoder()

categorical = df.select_dtypes(
    include="object"
).columns

for col in categorical:

    df[col] = le.fit_transform(
        df[col]
    )

print("\nEncoding Completed")


# =====================================================
# 13. FINAL VALIDATION
# Purpose: Verify cleaned dataset
# =====================================================

print("\n========== FINAL REPORT ==========\n")

print("Missing Values")

print(df.isnull().sum())

print("\nDuplicates")

print(df.duplicated().sum())

print("\nFinal Shape")

print(df.shape)

print("\nData Types")

print(df.dtypes)


# =====================================================
# 14. SAVE CLEANED DATASET
# Purpose: Export cleaned file
# =====================================================

df.to_csv(

    "hotel_booking_cleaned.csv",

    index=False

)

print("\nDataset Saved Successfully")
```
